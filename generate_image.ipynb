```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn import linear_model
from sklearn import svm
import joblib
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
from PIL import Image
import PIL
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv",na_values={
    'bioactivity_class':['intermediate']
})
df
```

```python
clean = df.fillna({
    'bioactivity_class':'active',
    'canonical_smiles': 'COc1ccc2nc3cccc(OC)c3nc2c1'
})
clean.dropna()
```

```python
cou = 0
for i in clean['canonical_smiles']:
    img = Chem.MolFromSmiles(i)
    get_img  = Chem.Draw.MolToImage(img)
    get_img.save(f"./image/img_{cou}.jpg")
    cou= cou+1
    
```
